```python
import numpy as np

n1 = np.array([1, 2, 3, 4, 5])
n2 = np.array([6, 7, 8, 9, 10])
n3 = n1 + n2
n4 = n3 * 10 / 5 ** 2
n4.tolist()
```

```python
import matplotlib.pyplot as plt
x = np.linspace(-3, 3, 50)
y = 2 * x + 1
z = x ** 2

figure = plt.figure()
axes = figure.add_subplot()
axes.plot(x, y, color='red', linewidth=1.0, linestyle='--')
axes.plot(x, z, color='blue', linewidth=1.5, linestyle='-')
plt.show()
```

```python
t = np.arange(0, 2.5, 0.1)
y1 = np.sin(np.pi * t)
y2 = np.cos(np.pi * t)

figure = plt.figure()
axes = figure.add_subplot()
axes.plot(t, y1 , "--")
axes.plot(t, y2 , "r-")
plt.show()
```
