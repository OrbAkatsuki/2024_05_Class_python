```python
import matplotlib.pyplot as plt

index = [3, 4, 5, 6, 7]
subject = ['apple', 'banana', 'orange', 'tomato', 'guava']
values = [5000, 6000, 7000, 8000, 9000]

figure = plt.figure()
axes = figure.add_subplot(1,1,1)
axes.bar(index, values)
axes.set_xticks(index,labels=subject)

for x,y in zip(index,values):
    axes.annotate(str(y),(x-0.2,y+200))

xmin, xmax, ymin, ymax = axes.axis()
xmin -= 0.2
xmax += 0.2
ymin = 3000
ymax += 1000
axes.axis(xmin=xmin,xmax=xmax,ymin=ymin,ymax=ymax)
plt.show()
```

```python
import matplotlib.pyplot as plt

index = [3, 4, 5, 6, 7]
subject = ['apple', 'banana', 'orange', 'tomato', 'guava']
values = [5700, 3200, 7050, 8000, 9600]

figure = plt.figure()
axes = figure.add_subplot(1,1,1)
axes.plot(index, values,'or-.')
axes.set_xticks(index,labels=subject)

for x,y in zip(index,values):
    axes.annotate(str(y),(x-0.2,y+200))

xmin, xmax, ymin, ymax = axes.axis()
xmin -= 0.2
xmax += 0.2
ymin = 3000
ymax += 1000
axes.axis(xmin=xmin,xmax=xmax,ymin=ymin,ymax=ymax)
plt.show()
```

```python
import matplotlib.pyplot as plt
st1 = {'chinese':78,'math':89,'english':80,'history':85,'science':88}
st2 = {'chinese':85,'math':75,'english':90,'history':88,'science':80}
subject = list(st1.keys())
values1 = list(st1.values())
values2 = list(st2.values())

figure = plt.figure(figsize=(10,5))
axes = figure.add_subplot(1,1,1)
axes.plot(subject,values1,'or-.')
axes.plot(subject,values2,'sb--')
plt.show()
```
